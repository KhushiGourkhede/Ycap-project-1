```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("file.txt", encoding="utf-8")
text_document = loader.load()

print(text_document)
```

```python
#text splitter 
from langchain_text_splitters import RecursiveCharacterTextSplitter
textSplitter = RecursiveCharacterTextSplitter(chunk_size = 200 , chunk_overlap=20)
finalDocument = textSplitter.split_documents(text_document)
finalDocument
```

```python
#embbeding using huggingface model 
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model="all-MiniLM-L6-v2")
```

```python
print(embeddings)
```

```python
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(finalDocument, embeddings)
db 
```

```python
pip install faiss-cpu
```

```python
query = input("Enter your question for c++ :")
docs = db.similarity_search(query)
for i in range(len(docs)):
    print(docs[i].page_content)
```
